```python
#univariate analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/HP/Desktop/Vignesh VIT files/Cs-python/titanic.csv')

# Explore the target variable "Survived"
survived_counts = df['survived'].value_counts()
survival_rate = survived_counts[1] / len(df) * 100
print(f"Survival Rate: {survival_rate:.2f}%")

# Plot the distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Distribution')
plt.show()

# Analyze categorical variables
categorical_vars = ['sex', 'pclass', 'sibsp']
for var in categorical_vars:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=var, hue='survived')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.title(f'Survival Distribution by {var}')
    plt.legend(['No', 'Yes'])
    plt.show()

# Analyze numerical variables
numerical_vars = ['age', 'fare']
for var in numerical_vars:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=var, kde=True, hue='survived')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.title(f'Distribution of {var} by Survival')
    plt.legend(['No', 'Yes'])
    plt.show()
```

```python
#bivariate analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/HP/Desktop/Vignesh VIT files/Cs-python/titanic.csv')

# Analyze the relationship between "Survived" and "Sex"
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='survived', hue='sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Distribution by Sex')
plt.legend(['Female', 'Male'])
plt.show()

# Analyze the relationship between "Survived" and "Pclass"
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='survived', hue='pclass')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Distribution by Passenger Class')
plt.legend(title='Pclass')
plt.show()

# Analyze the relationship between "Survived" and "Age"
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='age', hue='survived', kde=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age by Survival')
plt.legend(['No', 'Yes'])
plt.show()

# Analyze the relationship between "Survived" and "Fare"
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='fare', hue='survived', kde=True)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Distribution of Fare by Survival')
plt.legend(['No', 'Yes'])
plt.show()
```

```python
#multivariate analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/HP/Desktop/Vignesh VIT files/Cs-python/titanic.csv')

# Analyze the relationship between "Survived", "Sex", and "Pclass"
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='sex', y='survived', hue='pclass')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Sex and Passenger Class')
plt.legend(title='Pclass')
plt.show()

# Analyze the relationship between "Survived", "Age", and "Fare"
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='age', y='fare', hue='survived')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Survival Scatterplot: Age vs Fare')
plt.legend(title='Survived')
plt.show()

# Analyze the relationship between "Survived", "Embarked", and "Pclass"
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='embarked', y='pclass', hue='survived')
plt.xlabel('Embarked')
plt.ylabel('Passenger Class')
plt.title('Survival Boxplot: Embarked vs Pclass')
plt.legend(title='Survived')
plt.show()
```

```python
import pandas as pd

# Load the Titanic dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/HP/Desktop/Vignesh VIT files/Cs-python/titanic.csv')

# Check for missing values
print(df.isnull().sum())

# Dropping rows with missing values
df_dropna = df.dropna()
print("After dropping rows with missing values:")
print(df_dropna.isnull().sum())

# Impute missing values with mean
df_fillna_mean = df[['age','fare']].fillna(df.mean(numeric_only=True))
print("After imputing missing values with mean:")
print(df_fillna_mean.isnull().sum())

# Impute missing values with mode
df_fillna_mode = df.fillna(df.mode().iloc[0])
print("After imputing missing values with mode:")
print(df_fillna_mode.isnull().sum())
```

```python
import pandas as pd
import numpy as np

# Load the Titanic dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/HP/Desktop/Vignesh VIT files/Cs-python/titanic.csv')

# Detect outliers and replace them with the mean value
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    column = df[col]
    mean = column.mean()
    std = column.std()
    outliers = (column < mean - 3 * std) | (column > mean + 3 * std)
    df.loc[outliers, col] = mean

# Print the modified DataFrame
print(df.head())


```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the Titanic dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/HP/Desktop/Vignesh VIT files/Cs-python/titanic.csv')

# Check for categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_columns)

# Perform one-hot encoding using get_dummies()
df_encoded = pd.get_dummies(df, columns=categorical_columns)
print("Encoded DataFrame:")
print(df_encoded.head())

# Alternatively, perform label encoding using LabelEncoder
label_encoder = LabelEncoder()
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

print("Encoded DataFrame:")
print(df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the Titanic dataset into a pandas DataFrame
df = pd.read_csv('C:/Users/HP/Desktop/Vignesh VIT files/Cs-python/titanic.csv')

# Split the data into dependent (target) variable and independent (feature) variables
X = df.drop('survived', axis=1)  # Independent variables (features)

# Separate numeric and non-numeric columns
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns
non_numeric_cols = X.select_dtypes(include=['object']).columns

# Scale numeric columns using StandardScaler
scaler = StandardScaler()
X_scaled_numeric = pd.DataFrame(scaler.fit_transform(X[numeric_cols]), columns=numeric_cols)

# Concatenate the scaled numeric columns with the non-numeric columns
X_scaled = pd.concat([X_scaled_numeric, X[non_numeric_cols]], axis=1)

# Print the first few rows of the scaled DataFrame
print("Scaled DataFrame:")
print(X_scaled.head())
```
